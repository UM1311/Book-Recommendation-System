```python
rb=ratings.merge(books, on='ISBN')
```

```python
rb.head(3)
```

```python
number_rating = rb.groupby("Book-Title")['Book-Rating'].count().reset_index()
```

```python
number_rating.rename(columns={"Book-Rating":"No. of Ratings"}, inplace=True)
```

```python
number_rating.head(2)
```

```python
final_rating= rb.merge(number_rating, on="Book-Title")
```

```python
final_rating.head(2)
```

```python
final_rating=final_rating[final_rating['No. of Ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(["User-ID","Book-Title"],inplace =True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns="User-ID",index="Book-Title",values="Book-Rating")
```

```python
book_pivot.fillna(0,inplace = True)
```

```python
book_pivot
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm ="brute",n_neighbors=6)
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[240, : ].values.reshape(1,-1))
```
